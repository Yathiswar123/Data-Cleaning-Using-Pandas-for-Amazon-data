```python
import re
```

```python
def standardized_disc(disc_str):
    import re
    numeric_value = re.findall(r'-?/d+',disc_str)
    currency_value = re.findall(r'[A-Z]+',disc_str)
    if len(numeric_value) == 0:
        return None
    elif len(currency_value)==0:
        return float(numeric_value[0])
    else:
        return float(numeric_value[0])/100 if len(numeric_value) == 1 else None
```

```python
data['discount'].unique
```

```python
def percentage_symbol(discount):
    if '%' not in discount:
        return f"{discount}%"
    else:
        return discount
```

```python
data['discount_percentage'] = data['discount'].apply(percentage_symbol)
```

```python
exchange_rates = {'$':1,'USD':1,'INR': 00.12,'Gbp':1.26}
```

```python
def convert_To_usd(row):
    currency = row['currency']
    if currency in exchange_rates:
        return row['initial_price']*exchange_rates[currency],'USD'
    else:
        return row['initial_price'],currency
    
```

```python
df = pd.DataFrame(data)
```

```python
df['Price']= df.apply(convert_To_usd,axis=1)
```

```python
df.drop(columns = ['currency','initial_price'],inplace = True)
```

```python
def Usd_symbol(final_price):
    if '$' not in final_price:
        return f"${final_price}"
    else:
        return final_price
```

```python
df['Final_Price'] = df['final_price'].apply(Usd_symbol)
```
